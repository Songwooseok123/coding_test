# 특정 열 조건에 따라서 컬럼 만들거나 덮어씌우기

```python
def get_grade(x):
    if x >=10:
        return '매우 나쁨'
    elif 5<=x < 10:
        return "보통"
    else:
        return "좋음"
#df['windspeed'] = df['windspeed'].apply(get_grade)
df['특징'] = df['windspeed'].apply(get_grade)
df.head()

```

# 특정 열 value count하고, csv 파일로 저장하기 

```python
value_counts = df['특징'].value_counts().reset_index()
value_counts.columns = ['특징', '개수']
value_counts.to_csv('특징_통계.csv', index=False)
value_counts
```
