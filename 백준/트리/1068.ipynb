# 나의 풀이
- 또 틀림

```python
n= 5
a = [-1,0 ,0, 1, 1]
remove = 1


# 부모노드로 자식노드를 만들 수 있나? 
children = []
for i in range(n):
    i_child =[]
    for aa_index,aa in enumerate(a):
        if aa == i:
            i_child.append(aa_index)
    children.append(i_child)

#print(children)

def child_list(children,remove):
    ss =set()
    ss.add(remove)
    for i in children[remove]:
        ss.add(i)
        #print(children,i)
        child_list(children,i)
    return ss 

ss = child_list(children,remove)
for c_i,c in enumerate(children):
    if len(c)>0:
        ss.add(c_i)

if a[remove]== -1:
    print(0)
else:
    print(n -len(ss))
```

# 모범답안



0. 루트 노드 먼저 구하기 
1. 사라지는 노드에 색칠하기 -> 부모를 타고 올라가다가 remove를 만나면 사라지는 노드임. 
2. 자식이 있는 노드에 색칠하기 

검은색 먼저 색칠하고, 빨간색 색칠할 땐, 검은색 되어있으면 걍 넘어감 

```python
n= 5
parent = [-1,0 ,0, 1, 1]
remove = 2
```

```python

# 0.루트노드 찾기 
for parent_index,parent_value in enumerate(parent):
    if parent_value ==-1:
        root= parent_index 
        break
# 1.사라지는 노드 검은색으로 색칠
black = [0]*n

for i in range(n):
    
    u =i 
    while True:
        if u ==remove:
            black[i]=1
            break
        if u ==root:
            break 
        u = parent[u]
    

    #print(black)

# 2. 자식 있는 노드 빨간색으로 색칠(검은색은 그냥 넘어가고, 그렇지 않으면 부모 노드에 색칠 )

red = [0]*n
for i in range(n):
    if black[i] ==1:
        continue
    if i ==root:
        continue
        
    red[parent[i]]=1 # 자식이 있느 노드 
count = 0
for i in range(n):
    if black[i] ==0 and red[i] ==0:
        count +=1
print(count)
    
```
