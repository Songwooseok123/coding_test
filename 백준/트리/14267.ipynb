# 나의 답안
- 모르겠음

```python
# 일단 children 부터 구하고 시픙ㅁ

n,m = 5,3
parent = [-1,1,2,3,4]
#parent = [-1,1,1,2,2]
praise = [[2,2],[3,4],[5,6]]


children = []
for i in range(1,n+1):
    i_child=[]
    for p_index,p in enumerate(parent):
        if p ==i:
            i_child.append(p_index)
    children.append(i_child)


black = [0]*n

for i in range(n):
    if parent[i] == -1:
        root = i
        break
black[root] = 0
        
for praisement in praise:
    # [2,2]
    u, amount = praisement[0]-1, praisement[1] # 1,2
    while True:
        if len(children[u]) == 0:
            break
        for i in range(len(children[u])):
            black[u] +=amount
            if len(children[u]) == 0:
                break
            u = children[u][i]
                
```

# 모범답안 
1. parent 주어졌는데, index 1부터 시작하면, 인덱스 바꿔주기
2. 자신의 칭찬점수 구하기
3. 부모의 칭찬점수 더하기


```python
n,m = 5,3
parent = [-1,1,1,2,2]
praise = [[2,2],[3,4],[5,6]]
```

```python
# 1.  
for i in range(1,n):
    parent[i] -=1

# 2.
good = [0]*n

for praisement in praise :
    idd,score = praisement
    good[idd-1] += score
#print(good)

total_good = [0]*n

for i in range(1,n):
    total_good[i] = good[i]+total_good[parent[i]]
    
print(*total_good) 
```
