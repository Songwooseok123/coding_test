```python
v,e = list(map(int,input().split()))
k = int(input())-1

adj = [[] for i in range(v)]
for i in range(e):
    a,b,weight = list(map(int,input().split()))
    adj[a-1].append((b-1,weight))
    
```

```python
adj # id, 거리 
```

```python
from heapq import heappush,heappop

dist = [1e9]*v
dist[k] = 0

ding = []
heappush(ding,(0,k) ) # 거리 , id 

while ding:
    d,u = heappop(ding) # 옮길 거리, id 

    if d!= dist[u]: # 지금 꺼낸 노드의 거리가 dist[u] 랑 다르면, 이미 더 짦은 경로로 업데이트 된 상태-> 그래서 이건 오래된 정보라 무시 
        continue
    
    for id_,weight in adj[u]:  # [(1, 2), (2, 3)] # 인접한 애의 아이디와 거리 
        if dist[id_] > dist[u] +weight:
            dist[id_] = dist[u] +weight
            heappush(ding, (dist[u] +weight,id_))
for i in range(v):
    print(dist[i] if dist[i]!=1e9 else "INF")
```
