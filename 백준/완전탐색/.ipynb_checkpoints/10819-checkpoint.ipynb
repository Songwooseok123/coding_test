```python
from itertools import permutations
import sys
n = int(sys.stdin.readline())
ll= list(map(int,sys.stdin.readline().split()))
per_li  = list(permutations(ll,n))

def dd(a,b):
    return abs(a-b)
sums_list= []
for ll in per_li:
    sums=0
    for i in range(len(ll)-1):
        sums += dd(ll[i],ll[i+1])
    sums_list.append(sums)
print(max(sums_list))
```
