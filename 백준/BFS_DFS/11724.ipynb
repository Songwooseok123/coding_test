```python
# 어떤 시작점 = node 고르고 
# node 에서 갈 수 있는 노드들 다 모으고, 

# 전체 노드에서 위의 노드 리스트 빼고, 나머지 중에서 위의 과정 반복 하다가 전체 노드 다 가면 종료 

v,e = list(map(int,input().split()))


adj = [[] for i in range(v)]
for i in range(e):
    a,b = list(map(int,input().split()))
    adj[a-1].append(b-1)
    adj[b-1].append(a-1)
adj

# dfs
def dfs(node):
    visit[node]= True
    for j in adj[node]:
        if visit[j] == False:
            dfs(j)


visit = [False]*v
#dfs(0)

count = 0
for i in range(v):
    if visit[i] ==False:
        count +=1
        dfs(i)
print(count)


# bfs 

from collections import deque
visit = [False]*v

cc=0
for i in range(v):
    if visit[i] ==True:
        continue
    
    cc+=1
    node = i
    
    queue = deque()
    queue.append(node)
    visit[node] =True
    
    while queue:
        candidate = queue.popleft()
        for j in adj[candidate]:
            
            if visit[j] ==False:
                queue.append(j)
                visit[j] = True
    #print(i, visit)
print(cc)            
```
